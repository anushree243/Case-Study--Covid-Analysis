### Find Maximum infection rate for all of the countries.

```python
corona_df_agg.loc[['China']][0:1]
```

```python
corona_df_agg.loc[['Korea, South']][0:1]
```

```python
countries = list(corona_df_agg.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_df_agg.loc[c][1:].diff().max())

max_infection_rates
```

```python
corona_df_agg['max_infection_rates'] = max_infection_rates
```

```python
corona_df_agg.head()
```

```python
corona_df = pd.DataFrame(corona_df_agg['max_infection_rates'])
```

```python
corona_df.head()
```
